```python
#Round Robin
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:/Users/SAMARTH G VASIST/Downloads/TaskLog_RR.csv")

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:/Users/SAMARTH G VASIST/Downloads/TaskScheduleLog_RR.csv")


#print(df)


#worker 1 
w1_et=[]
w2_et=[]
w3_et=[]


for i in range(0,len(df)):
    if(df.loc[i,'Worker']==7000):
        w1_et.append(df.loc[i,'Scheduled_time'])
    
    elif(df.loc[i,'Worker']==7001):
        w2_et.append(df.loc[i,'Scheduled_time'])
    
    
    elif(df.loc[i,'Worker']==7002):
        w3_et.append(df.loc[i,'Scheduled_time'])



w1_et.sort()
w2_et.sort()
w3_et.sort()


final_time=max(w1_et[-1],w2_et[-1],w3_et[-1])
#print(final_time)
w1_et.append(final_time)
w2_et.append(final_time)
w3_et.append(final_time)

num_tasks=[i for i in range(0,len(w1_et))]
#print(num_tasks)
#print(num_tasks)
plt.plot(w1_et,num_tasks,linestyle='dashed',color='r',label="worker1")
plt.legend(loc="upper left")

num_tasks=[i for i in range(0,len(w2_et))]
#print(num_tasks)
#print(num_tasks)
plt.plot(w2_et,num_tasks,linestyle='dashdot',color='b',label="worker2")
plt.legend(loc="upper left")


num_tasks=[i for i in range(0,len(w3_et))]
#print(num_tasks)
#print(num_tasks)
plt.plot(w3_et,num_tasks,linestyle='solid',color='k',label="worker3")
plt.legend(loc="upper left")



plt.title('Round Robin Algorithm tasks Vs time ')
plt.xlabel('time')
plt.ylabel('num_tasks')

plt.show()
```
